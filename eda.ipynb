```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = 'airline_passenger_satisfaction.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
print("Shape of the dataset:", df.shape)
```

```python
print("\nSummary of the dataset:")
print(df.describe(include='all'))
```

```python
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
df = df.dropna()  # Example: Dropping rows with missing values
print("\nShape after handling missing values:", df.shape)
```

```python
print("\nData types of each column:")
print(df.dtypes)
```

```python
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
print("\nUnivariate Analysis:")
for column in df.select_dtypes(include=[np.number]).columns:
    print(f"\nColumn: {column}")
    print(df[column].describe())
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True, bins=20, color='blue')
    plt.title(f"Distribution of {column}")
    plt.show()
```

```python
print("\nBivariate Analysis:")
numerical_columns = df.select_dtypes(include=[np.number]).columns
categorical_columns = df.select_dtypes(include=['object']).columns

for num_col in numerical_columns:
    for cat_col in categorical_columns:
        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[cat_col], y=df[num_col])
        plt.title(f"{num_col} vs {cat_col}")
        plt.xticks(rotation=45)
        plt.show()
```

```python
print("\nMultivariate Analysis:")
numerical_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
df = df.fillna(df.mean())
```

```python
scaler = StandardScaler()
numerical_data = df.select_dtypes(include=[np.number])
scaled_data = scaler.fit_transform(numerical_data)
```

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=df['Cluster'], cmap='viridis', s=50)
plt.title('Clusters Visualized using PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
print("\nCluster Summary:")
print(df.groupby('Cluster').mean())
```
